```python
import matplotlib.pyplot as plt
x = ["Python" , "C" , "C++" , "Java"]
y = [90,70,60,80]
plt.xlabel("Languages" , fontsize = 15)
plt.ylabel("Numbers" , fontsize = 15)
plt.title("Bar Graph" , fontsize = 20)
c = ["y","b","m","g"]
plt.bar(x,y,width = 0.6,color = c, edgecolor = "black" , linewidth = 3 , label = "Popularity")
plt.legend()
plt.show()
```

```python
import numpy as np
x = ["Python" , "C" , "C++" , "Java"]
y = [90,70,60,80]
z = [50,40,30,20]

width = 0.4
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("Languages" , fontsize = 15)
plt.ylabel("Numbers" , fontsize = 15)
plt.title("Bar Graph" , fontsize = 20)

plt.bar(p,y,width,color = "r", edgecolor = "black"  , label = "Popularity")
plt.bar(p1,z,width,color = "y", edgecolor = "black" ,   label = "Popularity_1")
plt.xticks(p+width/2 , x , rotation = 10)
plt.legend()
plt.show()
```
